```python
import pandas as pd #load library - pd is just an alias. I used pd because it's short and literally abbreviates pandas.
import numpy as np #Load numpy
pd.set_option('display.float_format', '{:,.0f}'.format)# To display large float numbers in clean decimal format instead of scientific notation
ids=['A001','A002','A003','A004','A005'] #A reusable list for ids
```

```python
seniority_data={'ID':ids,'Designation Level':[2,2,3,2,3]} #Dictionary for seniority level dataframe
seniority_df=pd.DataFrame(seniority_data) #Converting dictionary to dataframe
seniority_df.to_csv('Seniority_Level_DataFrame.csv',index=False) #Saving as csv
seniority_df #Checking output
```

```python
employee_data={'ID':ids,'Name':['John Alter','Alice Luxembourg','Tom Sabestine','Nina Adgra','Amy Johny'],'Gender':['M','F','M','F','F'],'City':['Paris','London','Berlin','Newyork','Madrid'],'Age':[25,27,29,31,30]} #Dictionary for employee dataframe
employee_df=pd.DataFrame(employee_data) #Converting dictionary to dataframe
employee_df.to_csv('Employee_DataFrame.csv',index=False) #Saving as csv
employee_df #Checking output
```

```python
projects=['Project '+str(i) for i in range(1,15)] #Creating project column by list comprehension
stat=['Failed','Ongoing','Finished'] #List for status to avoid repeated writing
project_data={'ID':['A001','A002','A003','A004','A005','A002','A005','A003','A001','A003','A001','A004','A004','A005'],'Project':projects,'Cost':[1002000,2000000,4500000,5500000,np.nan,680000,400000,350000,np.nan,300000,2000000,1000000,3000000,200000],'Status':[stat[2],stat[1],stat[2],stat[1],stat[2],stat[0],stat[2],stat[0],stat[1],stat[2],stat[0],stat[1],stat[2],stat[2]]} #Dictonary for project dataframe
project_df=pd.DataFrame(project_data) #Converting dictionary to dataframe
project_df.to_csv('Project_DataFrame.csv',index=False) #Saving as csv
project_df #Checking output
```

```python
#Task 2
project_df=pd.read_csv('Project_DataFrame.csv') #Read from the saved csv
running_sum=0 #Variable for running sum
count=0 #Variable for number of non null values
for i in range(len(project_df)): #Iterating through rows
    if pd.isna(project_df.loc[i,'Cost']): # If cost is null 
        if count==0: # If the first value is null then count will be 0 only
            project_df.loc[i,'Cost']=0 #Hence the cost will be 0
        else: # If count is >0 
            running_avg =  running_sum/count # Running average is calculated by running sum / count of non null values till now
            project_df.loc[i,'Cost']=running_avg # Assigning running average to cost value which is null
    else:
        running_sum+=project_df.loc[i,'Cost'] # If cost value is not null then we add value to running sum
        count+=1 # Increment count of non null values
project_df.to_csv('Project_DataFrame.csv',index=False) #Saving as csv
project_df #Checking output
```

```python
#Task 3
employee_df=pd.read_csv('Employee_DataFrame.csv') #Read from the saved csv
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', n=1, expand=True)
employee_df.drop(columns=['Name'],inplace=True)
employee_df.to_csv('Employee_DataFrame.csv',index=False) #Saving as csv
employee_df #Checking output
```

```python
#Task 4
seniority_df=pd.read_csv('Seniority_Level_DataFrame.csv') #Read from the saved csv
employee_df=pd.read_csv('Employee_DataFrame.csv') #Read from the saved csv
project_df=pd.read_csv('Project_DataFrame.csv') #Read from the saved csv
temp=pd.merge(employee_df,seniority_df,on='ID',how='left') # left join employee and seniority levels for cases when employees don't have seniority level 
Final=pd.merge(temp,project_df,on='ID',how='left') # left join temp and project for cases when employees don't have projects
Final #Checking output
```

```python
#Task 5
Final['Bonus']=np.where(Final['Status']=='Finished',Final['Cost']*0.05,0) # 5% bonus to employees who have finished the project
Final #Checking output
```

```python
#Task 6
Final['Designation Level']=np.where(Final['Status']=='Failed',Final['Designation Level']+1,Final['Designation Level']) #Demoting designation level for each project which was failed
Final = Final[Final['Designation Level'] <= 4] # Removing designations above 4
Final #Checking output
```

```python
#Task 7
Final['First Name']=np.where(Final['Gender']=='M','Mr.' + Final['First Name'],'Mrs.'+Final['First Name']) # Adding Mr. and Mrs. based on gender
Final.drop(columns='Gender',inplace=True) #Dropping gender column
Final #Checking output
```

```python
#Task 8
for i in range(len(Final)): # For loop to iterate through Final dataframe
    if Final.loc[i,'Age']>29 and Final.loc[i,'Designation Level']>1: # If age > 29 and designation level is not 1 
        Final.loc[i,'Designation Level']-=1 # Then promote designation
Final #Checking output
        
```

```python
#Task 9
TotalProjCost = Final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index() #Groupby id and first name and sum the cost for each employee, reset index to convert to flat dataset
TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True) # Rename the column to total cost
TotalProjCost #Checking output
```
